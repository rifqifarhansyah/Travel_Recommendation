```python
nlp = spacy.blank("en")
db = DocBin()
i = 0
for item in data:
    doc = nlp.make_doc(item["sentence"])
    ents = []
    for ent in item["entities"]:
        span = doc.char_span(ent["start"], ent["end"], label=ent["entity"])
        if span is None:
            msg = f"Skipping entity [{ent["start"]}, {ent["end"]}, {ent["entity"]}] in the following text because the character span '{doc.text[ent["start"]:ent["end"]]}' does not align with token boundaries:\n\n{repr(item["sentence"])}\n"
            warnings.warn(msg)
        else:
            ents.append(span)
    print(i)
    i += 1
    print(ents)
    doc.ents = ents
    db.add(doc)
db.to_disk("./dev.spacy")
```

```python
import json

file_path = "test.json"

with open(file_path, "r") as file:
    data = json.load(file)
```

```python
import spacy
nlp = spacy.load("./output/model-best") 

fromacc = 0
toacc = 0
for el in data:
    doc = nlp(el["sentence"])

    for ent in doc.ents:    
        if ent.label_ == "FROM_CITY":
            if ent.text == el["from_city"]:
                fromacc += 1
        else:
            if ent.text == el["to_city"]:
                toacc += 1
            
print(fromacc)
print(toacc)
```
